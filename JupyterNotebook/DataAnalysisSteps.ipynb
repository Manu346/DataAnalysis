**Step 5**

- Data Quality checks or Data cleaning
    - Eg- If some columns have punctuations or unneccesary symbols or having categorical data in numerical data

- Identified '-' in phone number. It can make numerical data into categorical data so joining it

```python
a=[]
for i in df['phone_no']:
    a.append(''.join(i.split('-')))
df['phone_no']=a
```

```python
df['phone_no']=df['phone_no'].astype('int64')
Categorical_columns= df.select_dtypes(include=['object'])
Numerical_columns=df.select_dtypes(exclude='object')
```

**Step 8**

- Outlier Analysis

    - Boxplot
    - Treat the outliers

BoxPLot

```python
plt.figure(figsize=(30,19))
for i in range(1,len(Numerical_columns.columns)):
    plt.subplot(3,4,i)
    sns.boxplot(df[Numerical_columns.columns[i]],orient='h')
    plt.title(f'Boxplot for {Numerical_columns.columns[i]}')
plt.show()
```

Treating outliers

```python
# Numerical_columns.drop(columns='churn',inplace=True)
for i in range(1,len(Numerical_columns.columns)):
    q1=np.quantile(df[Numerical_columns.columns[i]],0.25)
    q3=np.quantile(df[Numerical_columns.columns[i]],0.75)
    iqr=q3-q1
    lower_bound=q1-1.5*iqr
    upper_bound=q3+1.5*iqr
    df[Numerical_columns.columns[i]]=df[Numerical_columns.columns[i]].apply(lambda x: upper_bound if x>upper_bound else lower_bound if x<lower_bound else x)
plt.figure(figsize=(30,19))
for i in range(1,len(Numerical_columns.columns)):
    plt.subplot(3,4,i)
    sns.boxplot(df[Numerical_columns.columns[i]],orient='h')
    plt.title(f'Boxplot for {Numerical_columns.columns[i]}')
plt.show()
```

# Feature Engineering

**Step 10**

- Convert categorical to numerical....Encoding
    - LabelEncoder
    - OneHotEncoder

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in Categorical_columns.columns:
    df[i]=le.fit_transform(df[i])
df
```

**Step 11**

- Scaling
    - Standardization
    - Normalization

```python
#standardization 
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
for i in Numerical_columns.columns:
    df[i]=sc.fit_transform(df[i].values.reshape(-1,1))
df
```

```python
# Normalization
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
for i in Numerical_columns.columns:
    df[i]=sc.fit_transform(df[i].values.reshape(-1,1))
df
```
